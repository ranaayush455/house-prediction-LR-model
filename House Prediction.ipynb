```python
import statsmodels.formula.api as smf
annova_model=smf.ols(formula='SalePrice~C(Date)', data=data_df)
annova_result=annova_model.fit()
print(annova_result.summary())
```

```python
data_df.drop(columns=['Street', 'Utilities', 'LandSlope', 'Date'],inplace=True)
```

```python
data_df
```

```python

from sklearn.model_selection import train_test_split
y=data_df['SalePrice']
data_df.drop(columns='SalePrice', inplace=True)
X_train, X_test, Y_train, Y_test=train_test_split(data_df,y,test_size=0.2,random_state=100)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for col in X_train.columns.values:
    if X_train[col].dtypes=='object':
        data=X_train[col].append(X_train[col])
        le.fit(data.values)
        X_train[col]=le.transform(X_train[col])
```

```python
X_train
```

```python
for col in X_test.columns.values:
    if X_test[col].dtypes=='object':
        datas=X_test[col].append(X_test[col])
        le.fit(datas.values)
        X_test[col]=le.transform(X_test[col])
X_test
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,Y_train)
model.predict(X_test)
model.coef_
```

```python
model.score(X_train, Y_train)
```

```python
model.score(X_test, Y_test)
```

```python
Predicted_values=model.predict(X_test)
Predicted_df=pd.DataFrame({'actual':Y_test, 'predicted':Predicted_values})
Predicted_df
```
